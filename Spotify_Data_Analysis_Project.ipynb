converting seconds to mili seconds to just seconds 

```python
# Convert duration in milliseconds to seconds and store it in a new column "duration"
df_tracks["duration"] = df_tracks["duration_ms"].apply(lambda x: round(x / 1000))

# Drop the "duration_ms" column from the DataFrame
df_tracks.drop("duration_ms", inplace=True, axis=1)
```

creating coorelation (heatmap) for 3 unwanted columns 

```python
# Drop the specified columns
corr_df = df_tracks.drop(["key", "mode", "explicit" ,"artists" , "id_artists" , "id" , "name"], axis=1).corr(method="pearson")

# Create a heatmap
plt.figure(figsize=(14, 6))
heatmap = sns.heatmap(corr_df, annot=True, fmt=".1g", vmin=-1, vmax=1, center=0, cmap="rocket", linewidths=0.5, linecolor="black")
heatmap.set_title("Correlation HeatMap Between Variables")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)

# Display the heatmap
plt.show()

```

```python
# Randomly sample a subset of rows from the DataFrame 'df_tracks'
sample_df = df_tracks.sample(int(0.004 * len(df_tracks)))

# Print the length (number of rows) of the sampled DataFrame to the console
print(len(sample_df))
```

```python
# Create a scatter plot with a regression line to visualize the correlation between "energy" and "loudness" in the DataFrame 'sample_df'
plt.figure(figsize=(10, 6))
sns.regplot(data=sample_df, x="energy", y="loudness", color="#FF0000")

# Set the title of the plot
plt.title("loudness vs energy correlation")
```

```python
# Create a scatter plot with a regression line to visualize the correlation between "acousticness" and "popularity" in the DataFrame 'sample_df'
plt.figure(figsize=(10, 6))
sns.regplot(data=sample_df, x="acousticness", y="popularity", color="orange")

# Set the title of the plot
plt.title("acousticness vs acousticness correlation")
```

```python
df_tracks['dates'] = df_tracks.index.get_level_values('release_date')
df_tracks.date = pd.to_datetime(df_tracks.dates)
years = df_tracks.dates.dt.year
```

```python
df_tracks.info()
```

```python
sns.displot(years , discrete = True , aspect = 2 , height = 5 , kind = "hist").set(title = "Number of song release in year")
```

```python
total_dr = df_tracks.duration
fig_dims = (18 , 7)
fig, ax = plt.subplots(figsize = fig_dims)
fig = sns.barplot(x = years , y = total_dr , ax= ax , errwidth = False).set(title = "Years vs Duration")
plt.xticks(rotation = 90)
```
